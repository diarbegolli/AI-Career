```python

url = 'https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv'

import pandas as pd

df = pd.read_csv(url)
#print x-amount of top rows
print(df.head(5))
#print x-amount of end rows
print(df.tail(5))
#print headers
print(df.columns)
#print specific coloumn
print(df['Name'][0:5])
#print specific row or location
print(df.iloc[2,1])
print(df.iloc[2])

#changing and adding data

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

df.head(5)

#saving the data as excel

df.to_excel('modified.xlsx')

#FILTERING data
#print a row if matches (IF)
df.loc[df['Type 1'] == "Grass"]

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

#reseting the index to start from 0
new_df = new_df.reset_index()

print(new_df)


df.loc[df['Type 1'].str.contains('Fire|Grass')]

#Conditional Changes

df.loc[df['Total'] > 500, ['Generation' , 'Legendary']] = ['Test 1' , 'Test 2']
df

df = pd.read_csv(url)

#Aggregate statistics (GROUP BY)


#gropby average
df.groupby(['Type 1']).mean(numeric_only=True).sort_values('Defense')

#gropby sum
df.groupby(['Type 1']).sum(numeric_only=True)

#gropby count
df.groupby(['Type 1']).count()
```
